```python
import pandas as pd
```

# 1. Read a .CSV file from URL with Pandas
   #### Target Website: https://www.football-data.co.uk/data.php

```python
#reading .csv file from website
bundesliga24 = pd.read_csv('https://www.football-data.co.uk/mmz4281/2425/D1.csv')
```

```python
#showing dataframe
bundesliga24
```

```python
#renaming columns
bundesliga24 = bundesliga24.rename(columns={'FTHG':'HomeGoals',
                                         'FTAG':'AwayGoals'})
```

```python
#showing dataframe
bundesliga24
```
